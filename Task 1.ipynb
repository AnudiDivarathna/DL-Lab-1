```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import pandas as pd

# Step 1: Create a random 4x4 array from an exponential distribution
array_4x4 = np.random.exponential(scale=1, size=(4, 4))
print("Random 4x4 array:")
print(array_4x4)

# Step 2: Create a random 100000x1 array from the same exponential distribution and visualize the distribution
array_100000x1 = np.random.exponential(scale=1, size=100000)

plt.hist(array_100000x1, density=True, bins=50, histtype="step", color="blue", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="green", label="Uniform")
plt.hist(np.random.normal(size=100000), density=True, bins=50, histtype="step", color="red", label="Normal")

plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()

# Step 3: Plot Z = X^2 + Y^2 in a 3D plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(0, 50)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('Z = X^2 + Y^2')
plt.show()

# Step 4: Calculate and visualize correlations using seaborn
data = pd.read_csv("seaborn tutorial.ipynb")

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data_subset = data[features]

pearson_corr = data_subset.corr(method='pearson')
spearman_corr = data_subset.corr(method='spearman')

plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', center=0)
plt.title("Pearson Correlation")

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', center=0)
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
